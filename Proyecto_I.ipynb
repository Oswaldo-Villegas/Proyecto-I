# **Ejercicio N°1:**
Escribir un programa que pregunte al usuario por las ventas de los 
últimos 5 días, y muestre por pantalla una serie con los datos de 
las ventas por producto: Nro de unidades y precio unitario, 
indexada por los días, antes y después de aplicarles un descuento 
del 5%.

**Solución**

```python
# Se obtienen las librerias.
import numpy as np 
import pandas as pd 
Inicio = int(input('Introduce la fecha en número del día de hoy: '))    # Con Input le pedimos al usuario que ingrese la fecha actual.                      
                                                                        # Posteriormente con int se convierte el valor ingresado en entero. 

ventas={}  # Construimos un Diccionario vacio, en donde se almacenara los datos ingresados.

for i in range(Inicio, Inicio-5,-1): # Se construye un bucle for, en el cual se especifica la variable y el elemento a recorrer,
                                     # este ultimo se contruye con la funciónn range, tiene un valor inicial, un valor final y la secuencia.

#### Cuerpo del Bucle
# Dentro del bucle se indica varias condicionales, lo que significa que si se cumple un criterio las lineas de codigo
# dentro la condicion se ejecutaran.
  if i < 0:        # Esta condicional indica que si la fecha en numero es menor a cero, pide al usuario que ingrese los datos y estos los almacenara 
                   # dentro del diccionario.
    Unidades = int(input('Introduce el número de unidades Vendidas en el día ' + str(i+30) + ': ')) # Ingresamos
    Precio = float(input('Introduce el precio del producto (precio unitario) : '))
    ventas[i+30] =  np.array(Unidades)*np.array(Precio) # Se almacena los valores ingresados dentro del diccionario, donde las keys(claves) son las 
                                                        # fechas y los valores es la multiplicación de dos vectores, el vector producto por el vector 
                                                        # precio untitario (ventas por producto).
  elif 0 < i <=30: # La segunda condicional se ejecutara siempre y cuando la fecha ingresada en numero se encuentre entre 1 y 30.
    Unidades = int(input('Introduce el número de unidades Vendidas en el día ' + str(i) + ': ')) # Ingresamos 
    Precio = float(input('Introduce el precio del producto (precio unitario): '))
    ventas[i] =  np.array(Unidades)*np.array(Precio) # Se almacena los valores como: fecha (keys) y 
                                                     # ventas por producto (valores).
  else:            # Finlmente en el ultimo condicional se muestra los valores que no cumplen los criterios anteriores. 
    print('Error de Fecha') # Imprimimos
    break;                  # Se agrega un rompimiendo (break), esto para que no se muestre el mensaje < Error de Fecha > cinco veces.

# Finalmente, se imprime el resultado, para ello se construye una serie con los diccionarios obtenidos. 
# donde el index corresponde a la fecha del día, y el vector de datos a las ventas por producto.
ventas = pd.Series(ventas) 
print('Ventas\n', ventas)                     # Imprimimos las ventas por producto.
print('Ventas con descuento\n', ventas*0.95)  # Se imprime las ventas con descuento al 5%.
```

# **Ejercicio 2**
Escribir un programa que reciba un diccionario con los ingresos 
mensuales de los clientes de una empresa, y devuelva una serie 
con el ingreso mínimo, máximo y media.

**Solución**

```python
# importando las librerias.
import pandas as pd

Numero = int(input('Ingresa el numero total de Clientes: ')) # Se le pide al usuario que ingrese el número total de clientles. 

Datos = {}  # Se genera un Diccionario vacio, en donde se alamacenaran los Datos ingresados.

### Cosntruimos varios condicionales.
if Numero <= 0:  # Esta condicional indica que si el numero de clientes es menor a cero,
                 # se le muestre al usuario un menaje en donde se especifique que no es posible efectuar ninguna operación. 
  print('Sorry !! ... con el dato ingresado no se puede hacer mucho')
else:            # En caso contrario se le indica al usuario que ingrese los datos del cliente (nombre y Ingreso Mensual)
  for i in range(0, Numero, 1): # Se genera un bucle para ingresar los datos al diccionario.
    Cliente = input('Introduce el nombre del cliente ' + str(i+1) + ': ')
    Ingreso = int(input('Introduce el Ingreso Mensual del Cliente: '))
    i = Cliente  
    Datos[i]=Ingreso # Se almacena los datos obtenidos cliente e Ingreso.

# Finalmente se genara una función que devuelve el minimo, máximo y promedio de los valores del diccionario obtenido.
def ingresos_mensuales(ingresos):
  ingresos = pd.Series(ingresos)  # Se convierte el diccionario en una serie, esto con la finalidad de manipular sencillamente los datos.
  ingresos = {'ingreso minimo':ingresos.min(), 'ingreso máximo':ingresos.max(), 'ingreso promedio':ingresos.mean()} # Obtenemos el minimo, maximo y promedio.
  return pd.Series(ingresos)      # Devolvemos el diccionario de Ingresos como una serie.

#ingresos = {'Juan':145, 'María':245, 'Pedro':567, 'Carmen': 789, 'Luis': 489}
print('\n Se muestra los clientes de la empresa y sus ingresos mensuales \n', Datos) # Se obtienen los datos ingresados.
print('\n Se obtiene algunos medidas\n', ingresos_mensuales(Datos))                  # Se obtienen los resultados.
```

# **Ejercicio 3**
Escribir programa que genere y muestre por pantalla un DataFrame 
con los datos de la tabla siguiente:

| Mes     | Ventas | Gastos |
| ------- | -----: | -----: |
| Enero   |  30500 |  22000 |
| Febrero |  35600 |  23400 |
| Marzo   |  28300 |  18100 |
| Abril   |  33900 |  35700 |
| Mayo    |  42500 |  32450 |

A continuación, generar nuevas columna:

1. Columna 1 = Gasto/Ventas*100.
2. Columna 2 = Si Ventas>= 3000, imprimir “Meta Superada”, 
caso contrario imprimir “Meta no Superada”.
3. Columna 3 = Si Gasto>Ventas, calcular el sobre gasto.
4. Obtener el total de las ventas de los últimos 3 meses
5. Obtener el total de gastos de febrero a abril

**Solución**

```python
# Se importan las librerias necesairias.
import pandas as pd

# Se genera la variable Datos como un diccionario, en donde las claves son: Mes, Ventas y Gatos,
# y los valores son los elementos corespondientes a dichas variables.
Datos = {'Mes':['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo'], 'Ventas':[30500,35600,28300,33900,42500],'Gastos':[22000,23400,18100,35700,32450]}
Datos = pd.DataFrame(Datos) # La variable Datos, se convierte en un Data.Frame e imprimimos.
Datos 
```

```python
# Se genera la Columna 1 = Gasto/Ventas*100.
Columna1 = Datos['Gastos']/Datos['Ventas']*100
```

```python
# Se genera la Columna 2 = Si Ventas>= 3000, imprimir “Meta Superada”, caso contrario imprimir “Meta no Superada”.
columna2={}
for i in list(Datos['Ventas']):
  if i >= 3000:
   columna2[i] = 'Meta Superada'
  else:
    columna2[i] = 'Meta no Superada'

pd.DataFrame(columna2.values(), columns=['Columna 2'])
```

```python
# se genera la Columna 3 = Si Gasto>Ventas, calcular el sobre gasto.

columna = {}
columna3 =  Datos['Ventas'] - Datos['Gastos']
for i in list(columna3):
  if i < 0:
    columna[i] = Datos['Gastos'][1] - Datos['Ventas'][1]
  else:
    columna[i]= 0

pd.DataFrame(columna.values(), columns=['Columna 3'])
```

```python
# Se obtiene el total de las ventas de los últimos 3 meses
Datos.loc[2:5,'Ventas'].sum()
```

```python
# Se obtiene el total de gastos de febrero a abril
Datos.loc[1:3,'Gastos'].sum()
```
